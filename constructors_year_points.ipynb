```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\matti\OneDrive\Desktop\Projects_Analysis\F1\constructors_year_nationality_points.csv')
df
```

```python
df1=df[df['name'].isin(['Mercedes','Ferrari','Williams','McLaren','Red Bull'])].groupby(['name','year'])['tot_points'].sum().reset_index().sort_values(by='tot_points',ascending=False)
df1
```

```python
sns.lineplot(x='year',y='tot_points',hue='name',data=df1)
plt.legend(title='f1',fontsize=5,title_fontsize=13)
```
