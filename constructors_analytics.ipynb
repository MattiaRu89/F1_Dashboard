```python
columns_interval=df[['stop1', 'stop2', 'stop3', 'stop4',
       'stop5']]
columns_interval=columns_interval.astype('timedelta64[s]')
df[['stop1', 'stop2', 'stop3', 'stop4',
       'stop5']]=columns_interval
df.info()
df5=df
```

```python
df5.apply(lambda x:x.fillna(pd.Timedelta(0)),axis=0)
```

```python
df0=df5.groupby(['raceid', 'constructorid', 'constructors', 'nationality', 'year', 'gp',
       'points', 'positiontext', 'wins', 'fastestlaptime', 'fastestlapspeed',
       'quali_position', 'q1', 'q2', 'q3'])[['stop1', 'stop2', 'stop3', 'stop4',
       'stop5']].mean().reset_index()
df0
```

```python
df0.apply(lambda x:x.fillna(0),axis=0)
df0
```

```python
cols1=['stop1', 'stop2', 'stop3', 'stop4','stop5']
df0[cols1]=df0[cols1].fillna(pd.Timedelta(0))
df0
```

```python
df6=df0.groupby(['raceid', 'constructorid', 'constructors', 'nationality', 'year', 'gp',
       'points', 'positiontext', 'wins', 'fastestlaptime', 'fastestlapspeed',
       'quali_position', 'q1', 'q2', 'q3'])[['stop1', 'stop2', 'stop3', 'stop4',
       'stop5']].mean().reset_index()
df6
```

```python
cols=['stop1', 'stop2', 'stop3', 'stop4','stop5']
df6[cols]=df6[cols].apply(lambda x:x.dt.components.hours.astype(str).str.zfill(2)+':'+ 
                                     x.dt.components.minutes.astype(str).str.zfill(2) + ':' +
                                        x.dt.components.seconds.astype(str).str.zfill(2) +
                                        '.' + x.dt.components.milliseconds.astype(str).str.zfill(3), 
                                        axis=0)
df6
```

```python
df7=df6.drop_duplicates(subset=['raceid', 'constructorid', 'constructors', 'nationality', 'year', 'gp',
       'points', 'positiontext', 'wins'])
df7
```
